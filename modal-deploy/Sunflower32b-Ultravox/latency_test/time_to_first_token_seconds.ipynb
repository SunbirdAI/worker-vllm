```python
import pandas as pd
import numpy as np
import re
```

```python
# Load the content from your file
def time_to_first_token_seconds(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    
    timestamp = []
    log = []
    
    for line in lines:
        line = line.strip()
        if "Dec 17" in line:
            timestamp.append(line)
        else:
            log.append(line)
    
    df = pd.DataFrame({"timestamp": timestamp, "log": log})

    keywords = "Received request|Generated response"
    df = df[df['log'].str.contains(keywords, na=False)].reset_index(drop=True)
    conditions = [
        df['log'].str.contains("Received request", na=False),
        df['log'].str.contains("Generated response", na=False)
    ]
    choices = ['received', 'generated']
    df['type'] = np.select(conditions, choices, default="others")
    df = df[df.type.isin(choices)]

    df['dt'] = pd.to_datetime(df['timestamp'], format='%b %d %H:%M:%S.%f')
    df['time_to_first_token_seconds'] = (df['dt'].shift(-1) - df['dt']).dt.total_seconds()
    df = df[df['type'] == 'received'][['timestamp', 'type', 'time_to_first_token_seconds', 'log']]

    return df
```
